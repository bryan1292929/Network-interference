```python
# Code developed with AI assistance and reviewed by Bryan Park.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from tqdm.auto import tqdm

SEED = 519
N = 1000
n_sims = 10000
pi = 0.5
rho_sparse_factor = N ** (-0.3)

np.random.seed(SEED)

# Generates potential outcomes for latent u and exposure (w,r).
def get_potential_outcome(u, w, r):
    alpha = u
    tau = 1.0 + 4.0 * u
    beta = 2.0 + 2.0 * u
    gamma = 5.0
    delta = 4.0
    return alpha + tau * w + beta * r + gamma * r**2 + delta * w * r


# The HT estimator.
def horvitz_thompson(adj, w, u, pi):
    degrees = np.maximum(adj.sum(axis=1), 1)
    r_vector = (adj @ w) / degrees
    y_obs = get_potential_outcome(u, w, r_vector)
    weights = (w / pi) - ((1 - w) / (1 - pi))
    return np.mean(weights * y_obs)


# The Hajek estimator.
def hajek(adj, w, u):
    degrees = np.maximum(adj.sum(axis=1), 1)
    r_vector = (adj @ w) / degrees
    y_obs = get_potential_outcome(u, w, r_vector)

    n_treat = np.sum(w)
    n_control = np.sum(1 - w)

    if n_treat == 0 or n_control == 0:
        return np.nan

    treated_mean = np.sum(w * y_obs) / n_treat
    control_mean = np.sum((1 - w) * y_obs) / n_control
    return treated_mean - control_mean


# Computes limiting variances via Monte Carlo.
def calculate_limit_variances(pi, n_theory=10_000_000):
    u = np.random.rand(n_theory)

    y1 = get_potential_outcome(u, 1, pi)
    y0 = get_potential_outcome(u, 0, pi)

    R = (y1 / pi) + (y0 / (1 - pi))
    Q = -4.0 * np.log(np.maximum(1 - u, 1e-12))
    Z = R + Q

    V_HT = pi * (1 - pi) * np.mean(Z**2)
    V_HAJ = pi * (1 - pi) * (np.var(Z) + np.mean(Q) ** 2)

    return V_HT, V_HAJ


# Generates the half-graph.
def generate_deterministic_setup(n):
    u = np.arange(1, n + 1) / n

    i, j = np.indices((n, n)) + 1
    adj = ((i + j) > n).astype(float)
    np.fill_diagonal(adj, 0)

    return adj, u


# Randomly thins the half-graph to density rho.
def generate_deterministic_sparse_setup(n, rho):
    adj_dense, u = generate_deterministic_setup(n)

    mask = np.random.rand(n, n) <= rho
    mask = np.triu(mask, 1)
    mask = mask + mask.T

    return adj_dense * mask, u


# Draws a random graph from the limiting graphon at density rho.
def generate_random_graphon_setup(n, rho):
    u = np.random.rand(n)

    U_col = u[:, None]
    U_row = u[None, :]

    W_mat = (U_col + U_row > 1.0).astype(float)
    P_edge = rho * W_mat

    adj = (np.random.rand(n, n) <= P_edge).astype(float)
    adj = np.triu(adj, 1)
    adj = adj + adj.T

    return adj, u

# Helper functions.

def clean_array(x):
    x = np.asarray(x, dtype=float)
    return x[~np.isnan(x)]


def empirical_variance(x):
    x = clean_array(x)
    return np.var(x, ddof=1)


def normal_density(x, mean, var):
    sd = np.sqrt(var)
    return np.exp(-0.5 * ((x - mean) / sd) ** 2) / (sd * np.sqrt(2 * np.pi))


def qq_r_squared(data):
    data = clean_array(data)

    if len(data) < 3:
        return np.nan

    _, fit_params = stats.probplot(data, dist="norm", fit=True)
    return fit_params[2] ** 2
```

```python
V_HT_limit, V_HAJ_limit = calculate_limit_variances(pi)

print(f"HT:{V_HT_limit}")
print(f"HAJ:{V_HAJ_limit}")

results = {
    "Dense": {
        "Det": {"HT": [], "Hajek": []},
        "Fixed": {"HT": [], "Hajek": []},
        "Super": {"HT": [], "Hajek": []},
    },
    "Sparse": {
        "Det": {"HT": [], "Hajek": []},
        "Fixed": {"HT": [], "Hajek": []},
        "Super": {"HT": [], "Hajek": []},
    },
}

rho_d = 1.0
rho_s = rho_sparse_factor

# Runs six experiments (3 settings, dense or sparse).

adj_det, u_det = generate_deterministic_setup(N)
for _ in tqdm(range(n_sims), desc="Dense Det"):
    w = np.random.binomial(1, pi, N)
    results["Dense"]["Det"]["HT"].append(horvitz_thompson(adj_det, w, u_det, pi))
    results["Dense"]["Det"]["Hajek"].append(hajek(adj_det, w, u_det))

adj_fixed, u_fixed = generate_random_graphon_setup(N, rho_d)
for _ in tqdm(range(n_sims), desc="Dense Fixed"):
    w = np.random.binomial(1, pi, N)
    results["Dense"]["Fixed"]["HT"].append(horvitz_thompson(adj_fixed, w, u_fixed, pi))
    results["Dense"]["Fixed"]["Hajek"].append(hajek(adj_fixed, w, u_fixed))

for _ in tqdm(range(n_sims), desc="Dense Super"):
    adj_super, u_super = generate_random_graphon_setup(N, rho_d)
    w = np.random.binomial(1, pi, N)
    results["Dense"]["Super"]["HT"].append(horvitz_thompson(adj_super, w, u_super, pi))
    results["Dense"]["Super"]["Hajek"].append(hajek(adj_super, w, u_super))

adj_det_s, u_det_s = generate_deterministic_sparse_setup(N, rho_s)
for _ in tqdm(range(n_sims), desc="Sparse Det"):
    w = np.random.binomial(1, pi, N)
    results["Sparse"]["Det"]["HT"].append(horvitz_thompson(adj_det_s, w, u_det_s, pi))
    results["Sparse"]["Det"]["Hajek"].append(hajek(adj_det_s, w, u_det_s))

adj_fixed_s, u_fixed_s = generate_random_graphon_setup(N, rho_s)
for _ in tqdm(range(n_sims), desc="Sparse Fixed"):
    w = np.random.binomial(1, pi, N)
    results["Sparse"]["Fixed"]["HT"].append(horvitz_thompson(adj_fixed_s, w, u_fixed_s, pi))
    results["Sparse"]["Fixed"]["Hajek"].append(hajek(adj_fixed_s, w, u_fixed_s))

for _ in tqdm(range(n_sims), desc="Sparse Super"):
    adj_super_s, u_super_s = generate_random_graphon_setup(N, rho_s)
    w = np.random.binomial(1, pi, N)
    results["Sparse"]["Super"]["HT"].append(horvitz_thompson(adj_super_s, w, u_super_s, pi))
    results["Sparse"]["Super"]["Hajek"].append(hajek(adj_super_s, w, u_super_s))
```

```python
# This section creates plots of the simulation results.

plt.rcParams.update(
    {
        "axes.grid": True,
        "grid.alpha": 0.35,
        "grid.linestyle": "-",
        "axes.spines.top": False,
        "axes.spines.right": False,
        "font.size": 11,
    }
)

regimes = ["Dense", "Sparse"]
settings = [
    ("Det", "Deterministic Rule"),
    ("Fixed", "Sample Once"),
    ("Super", "Sample Always"),
]

BINS = 40

HT_HIST_COLOR = "tab:blue"
HAJ_HIST_COLOR = "tab:orange"
HT_LIMIT_COLOR = "tab:green"
HAJ_LIMIT_COLOR = "tab:red"

all_data = []
for regime in regimes:
    for setting_key, _ in settings:
        all_data.append(clean_array(results[regime][setting_key]["HT"]))
        all_data.append(clean_array(results[regime][setting_key]["Hajek"]))

all_data = np.concatenate(all_data)
global_min = np.min(all_data)
global_max = np.max(all_data)
x_buffer = 0.05 * (global_max - global_min)

xlims = (global_min - x_buffer, global_max + x_buffer)
global_bins = np.linspace(xlims[0], xlims[1], BINS + 1)
x_grid = np.linspace(xlims[0], xlims[1], 500)

fig, axes = plt.subplots(
    2,
    3,
    figsize=(18, 10),
    sharex=True,
    sharey=True,
)

for row_idx, regime in enumerate(regimes):
    for col_idx, (setting_key, setting_title) in enumerate(settings):
        ax = axes[row_idx, col_idx]

        ht_data = clean_array(results[regime][setting_key]["HT"])
        haj_data = clean_array(results[regime][setting_key]["Hajek"])

        ht_mean = np.mean(ht_data)
        haj_mean = np.mean(haj_data)

        ht_var = empirical_variance(ht_data)
        haj_var = empirical_variance(haj_data)

        ht_r2 = qq_r_squared(ht_data)
        haj_r2 = qq_r_squared(haj_data)

        ht_lim_var = V_HT_limit / N
        haj_lim_var = V_HAJ_limit / N

        ax.hist(
            ht_data,
            bins=global_bins,
            density=True,
            alpha=0.3,
            color=HT_HIST_COLOR,
            edgecolor=HT_HIST_COLOR,
            linewidth=0.85,
            label="HT simulation",
            zorder=1,
        )

        ax.hist(
            haj_data,
            bins=global_bins,
            density=True,
            alpha=0.3,
            color=HAJ_HIST_COLOR,
            edgecolor=HAJ_HIST_COLOR,
            linewidth=1.35,
            label="Hájek simulation",
            zorder=2,
        )

        ax.plot(
            x_grid,
            normal_density(x_grid, ht_mean, ht_lim_var),
            linestyle="--",
            linewidth=2.3,
            color=HT_LIMIT_COLOR,
            label="HT limit",
            zorder=3,
        )

        ax.plot(
            x_grid,
            normal_density(x_grid, haj_mean, haj_lim_var),
            linestyle="-",
            linewidth=2.2,
            color=HAJ_LIMIT_COLOR,
            label="Hájek limit",
            zorder=4,
        )

        stats_text = "\n".join(
            [
                f"HT_var  = {ht_var:.4f}",
                f"HT_R2   = {ht_r2:.3f}",
                f"HAJ_var = {haj_var:.4f}",
                f"HAJ_R2  = {haj_r2:.3f}",
            ]
        )

        ax.text(
            0.035,
            0.955,
            stats_text,
            transform=ax.transAxes,
            va="top",
            ha="left",
            fontsize=14,
            family="monospace",
            bbox=dict(
                boxstyle="round,pad=0.28",
                facecolor="white",
                edgecolor="0.75",
                linewidth=1.0,
                alpha=0.93,
            ),
            zorder=10,
        )

        ax.set_title(
            f"{regime} - {setting_title}",
            fontsize=19,
            fontweight="bold",
        )
        

        ax.set_xlim(xlims)

        if row_idx == 1:
            ax.set_xlabel(r"Estimate ($\hat{\tau}$)", fontsize=13)

        if col_idx == 0:
            ax.set_ylabel("Density", fontsize=13)

y_max = max(ax.get_ylim()[1] for ax in axes.flat)
for ax in axes.flat:
    ax.set_ylim(0, y_max)

handles, labels = axes[0, 0].get_legend_handles_labels()
unique = dict(zip(labels, handles))

fig.legend(
    unique.values(),
    unique.keys(),
    loc="upper center",
    ncol=4,
    frameon=False,
    fontsize=19,
    bbox_to_anchor=(0.5, 1.07),
)


plt.tight_layout()

plt.savefig("combined_ht_hajek_all_six_regimes.pdf", bbox_inches="tight")

plt.show()
```
